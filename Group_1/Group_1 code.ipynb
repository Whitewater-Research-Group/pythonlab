```python
import pytest
import cmath

def solve_quadratic(a, b, c):
    if a == 0:
        raise ValueError("Cofficient 'a' can not be 0")
    d = b**2 - 4*a*c
    root1 = (-b + cmath.sqrt(d)) / (2*a)
    root2 = (-b - cmath.sqrt(d)) / (2*a)
    return d, root1, root2

def test_real_roots():
    a, b, c = 1, -3, 2  # roots: 2 and 1
    d, root1, root2 = solve_quadratic(a, b, c)
    assert d == 1
    assert root1.real == 2
    assert root2.real == 1
    assert root1.imag == 0
    assert root2.imag == 0

def test_complex_roots():
    a, b, c = 1, 2, 3  # roots: complex
    d, root1, root2 = solve_quadratic(a, b, c)
    assert d == -8
    assert root1 == complex(-1, cmath.sqrt(8).real/2)
    assert root2 == complex(-1, -cmath.sqrt(8).real/2)

def test_a_zero():
    with pytest.raises(ValueError):
        solve_quadratic(0, 2, 3)
```
